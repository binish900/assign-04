```python
print("05_get_first_elemen")

import math  # Import the math library so we can use the sqrt function
import random
import pygame
import sys

def add_many_numbers(numbers)-> int:
    """
    Takes in a list of numbers and returns the sum of those numbers.
    """

    total_so_far: int = 0
    for number in numbers:
        total_so_far += number

    return total_so_far



def main():
    # Get the two side lengths from the user and cast them to be numbers
    ab: float = float(input("Enter the length of AB: "))
    ac: float = float(input("Enter the length of AC: "))

    # Calculate the hypotenuse using the two sides and print it out
    bc: float = math.sqrt(ab**2 + ac**2)
    print("The length of BC (the hypotenuse) is: " + str(bc))

    # Get the numbers we want to divide
    dividend: int = int(input("Please enter an integer to be divided: "))
    divisor: int = int(input("Please enter an integer to divide by: "))

    quotient: int = dividend // divisor  # Divide with no remainder/decimals (integer division)
    remainder: int = dividend % divisor  # Get the remainder of the division (modulo)

    print("The result of this division is " + str(quotient) + " with a remainder of " + str(remainder))#problem statement:  Simulate rolling two dice, and prints results of each roll as well as the total.
    # Import the random library which lets us simulate random things like dice!


    # Number of sides on each die to roll
    NUM_SIDES: int = 6

    # Roll die
    die1: int = random.randint(1, NUM_SIDES)
    die2: int = random.randint(1, NUM_SIDES)

    # Get their total
    total: int = die1 + die2

    # Print out the results
    print("Dice have", NUM_SIDES, "sides each.")
    print("First die:", die1)
    print("Second die:", die2)
    print("Total of two dice:", total)

    # Useful constants to help make the math easier and cleaner!
    DAYS_PER_YEAR: int = 365
    HOURS_PER_DAY: int = 24
    MIN_PER_HOUR: int = 60
    SEC_PER_MIN: int = 60

    # We can get the number of seconds per year by multiplying the handy constants above!
    print("There are " + str(DAYS_PER_YEAR * HOURS_PER_DAY * MIN_PER_HOUR * SEC_PER_MIN) + " seconds in a year!")

# Useful constants for this exercise
    SENTENCE_START: str = "Panaversity is fun. I learned to program and used Python to make my " # adjective noun verb

    # Get the three inputs from the user to make the adlib
    adjective: str = input("Please type an adjective and press enter. ")
    noun: str = input("Please type a noun and press enter. ")
    verb: str = input("Please type a verb and press enter. ")

    # Join the inputs together with the sentence starter
    print(SENTENCE_START + adjective + " " + noun + " " + verb + "!")

    numbers: list[int] = [1, 2, 3, 4, 5]  # Make a list of numbers
    sum_of_numbers: int = add_many_numbers(numbers)  # Find the sum of the list
    print("Sum of numbers:",sum_of_numbers)  # Print out the sum above

    # Problem Statement:
    # Write a program that doubles each element in a list of numbers.
    # For example, if you start with this list:
    # numbers = [1, 2, 3, 4]
    # You should end with this list:
    # numbers = [2, 4, 6, 8]

    def double_numbers(numbers):
        """Doubles each number in the input list.

        Args:
            numbers: A list of numbers.

        Returns:
            A new list with each number doubled.
        """
        doubled_numbers = []
        for number in numbers:
            doubled_numbers.append(number * 2)  # Multiply each number by 2 and add to the new list
        return doubled_numbers

    # Example usage:
    numbers = [1, 2, 3, 4]
    doubled_numbers = double_numbers(numbers)
    print("Original list:", numbers)
    print("Doubled list:", doubled_numbers)

# Initialize pygame
    pygame.init()

    # Constants for the grid and colors
    CELL_SIZE = 20
    GRID_WIDTH = 30
    GRID_HEIGHT = 20
    BLUE = (0, 0, 255)
    WHITE = (255, 255, 255)
    SCREEN_WIDTH = CELL_SIZE * GRID_WIDTH
    SCREEN_HEIGHT = CELL_SIZE * GRID_HEIGHT

    # Create the screen
    screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
    pygame.display.set_caption("Eraser Canvas")

    # Create the grid
    grid = [[BLUE for _ in range(GRID_WIDTH)] for _ in range(GRID_HEIGHT)]

    # Initial position of the eraser
    eraser_x = 0
    eraser_y = 0
    eraser_size = 3  # Make the eraser 3x3 cells


    def draw_grid():
        """Draws the grid of cells on the screen."""
        for y, row in enumerate(grid):
            for x, color in enumerate(row):
                pygame.draw.rect(screen, color, (x * CELL_SIZE, y * CELL_SIZE, CELL_SIZE, CELL_SIZE))
        # Draw the eraser rectangle
        pygame.draw.rect(screen, WHITE, (eraser_x * CELL_SIZE, eraser_y * CELL_SIZE, eraser_size * CELL_SIZE, eraser_size * CELL_SIZE))



    def handle_mouse_movement():
        """Handles mouse movement for the eraser."""
        global eraser_x, eraser_y
        mouse_buttons = pygame.mouse.get_pressed()
        if mouse_buttons[0]:  # Left mouse button is pressed (dragging)
            # Get the mouse position
            x, y = pygame.mouse.get_pos()
            # Calculate the grid coordinates of the mouse
            eraser_x = x // CELL_SIZE
            eraser_y = y // CELL_SIZE

            # Ensure the eraser stays within the grid bounds
            eraser_x = max(0, min(eraser_x, GRID_WIDTH - eraser_size))
            eraser_y = max(0, min(eraser_y, GRID_HEIGHT - eraser_size))

            # "Erase" the cells under the eraser
            for i in range(eraser_x, eraser_x + eraser_size):
                for j in range(eraser_y, eraser_y + eraser_size):
                    if 0 <= i < GRID_WIDTH and 0 <= j < GRID_HEIGHT: #check if the index is within the grid
                        grid[j][i] = WHITE



    # Main function to run the game.
    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            elif event.type == pygame.MOUSEMOTION:
                handle_mouse_movement()

        screen.fill(WHITE)  # Clear the screen
        draw_grid()  # Draw the grid
        pygame.display.flip()  # Update the display

    pygame.quit()
    sys.exit()

def add_three_copies(my_list, data):
    for i in range(3):
        my_list.append(data)


if __name__ == "__main__":
    message = input("Enter a message to copy: ")
    my_list = []
    print("List before:", my_list)
    add_three_copies(my_list, message)
    print("List after:", my_list)



def get_first_element(lst):
    """    Prints the first element of a provided list.    """

    print(lst[0])# There is no need to edit code beyond this point

def get_lst():
    """    Prompts the user to enter one element of the list at a time and returns the resulting list.    """
    lst = []
    elem: str = input("Please enter an element of the list or press enter to stop. ")
    while elem != "":
        lst.append(elem)
        elem = input("Please enter an element of the list or press enter to stop. ")
    return lst

def main():
    lst = get_lst()
    get_first_element(lst)


if __name__ == '__main__':
    main()
```
